```python
student = pd.read_csv('StudentsPerformance.csv')
```

```python
student.head()
```

```python
# 1
math_avg = student['math score'].mean()
read_avg = student['reading score'].mean()
write_avg = student['writing score'].mean()

math_avg,read_avg,write_avg
```

```python
# 2
student.groupby('gender')['writing score'].mean()
```

```python
# 3
student['total score'] = student[['math score', 'reading score', 'writing score']].sum(axis=1)
student['percentage'] = student['total score'] / 3

student
```

```python
# 4

import matplotlib.pyplot as plt

avg_scores = student.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
avg_scores.plot(kind='bar', figsize=(10,8))
plt.title("Average Scores by Parental Level of Education")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 5
student[student['percentage'] > 90].head(5)
```
